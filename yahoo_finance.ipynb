```python
import yfinance as yf 
from yahoo_fin.stock_info import get_data
import matplotlib.pyplot as plt 
import seaborn

stocks_1 = ['goog', 'amzn']
stock_2 = ['aapl', 'fb']

stock_data= yf.download(stocks_1, start="2019-01-01", end="2020-12-15", set_index='date')
#print(data.head(5))
closedStocks = stock_data.loc[:, "Close"].copy()
volumeStocks = stock_data.loc[:, "Volume"].copy()


top_plt = plt.subplot2grid((5,4), (0, 0), rowspan=3, colspan=4)
top_plt.plot(stock_data.index, closedStocks)
plt.title('Historical stock prices of Google and Amazon [01-04-2020 to 30-09-2020]')
bottom_plt = plt.subplot2grid((5,4), (3,0), rowspan=1, colspan=4)
bottom_plt.bar(stock_data.index, volumeStocks)
plt.title('\nAlphabet Inc. Trading Volume', y=-0.60)
plt.gcf().set_size_inches(12,8)
```
